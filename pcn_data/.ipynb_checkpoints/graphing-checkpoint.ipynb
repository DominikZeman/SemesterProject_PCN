# Predictive Coding Network Accuracy and Loss Results

### Intro
This code segment is for graphing the loss and accuracy of variations of the Predictive Coding Network as proposed and implemented 
by Han et al. 2018. (Paper: https://arxiv.org/pdf/1805.07526, Code: https://github.com/libilab/PCN-with-Local-Recurrent-Processing?tab=readme-ov-file) The code graphs various experiments that were conducted, they are ordered as follows:

### Circles Test:
The code implements the predictive coding architecture by implementing local recursion in adjacent layers through feedforward and feedback propagation. The number of times this happens in the code by Han et al. is 5. We wanted to find out the impact of the number of circles on the performance of the models. The first set of graphs is testing various numbers of circles. WE see that the performance increases slightly with the increase in circles until five circles, after which the performance decreases radically. The next three sets of graphs show the accuracies around the learning rate changes. Lastly, the final accuracies of the training and testing data are shown.

### Dataset test
The original paper tests the PCN on two datasets: ImageNet and CIFAR. The previous experiments we conducted were done on the CIFAR dataset. We also wanted to understand the implications of the PCN architecture on more task-oriented data, which could potentially combine proprioception with vision for future robotics use. For this we wanted to utilise the Human Action Recognition (HAR) dataset (https://www.kaggle.com/datasets/meetnagadia/human-action-recognition-har-dataset)  to find out whether there are any advantages of this architecture in such tasks. We trained the network on the HAR dataset with two different circle counts, 5 and 10 (15 circles was not able to run due to compute limitations).

### Network Size Test
The final tests we did were experimenting with the size of the network through the alteration of the number of hidden layers. The original network had 12 hidden layers and so we tested the performance with 8 hidden layers. We also trained three models with different circle counts (5, 10, and 15).

```python
# Number of circlers experiment

import pandas as pd
import matplotlib.pyplot as plt

# Function to extract numeric value from tensor-like string
def extract_numeric_value(tensor_str):
    return float(str(tensor_str).replace('tensor(', '').replace(')', ''))

# List of model files
model_files = [
    'training_metrics_PredNetBpD_1CLS_FalseNes_0.001WD_FalseTIED_6REP.csv',
    'training_metrics_PredNetBpD_2CLS_FalseNes_0.001WD_FalseTIED_1REP.csv',
    'training_metrics_PredNetBpD_3CLS_FalseNes_0.001WD_FalseTIED_1REP.csv',
    'training_metrics_PredNetBpD_4CLS_FalseNes_0.001WD_FalseTIED_1REP.csv',
    'training_metrics_PredNetBpD_5CLS_FalseNes_0.001WD_FalseTIED_6REP.csv',
    'training_metrics_PredNetBpD_6CLS_FalseNes_0.001WD_FalseTIED_1REP.csv',
    'training_metrics_PredNetBpD_7CLS_FalseNes_0.001WD_FalseTIED_1REP.csv',
    'training_metrics_PredNetBpD_8CLS_FalseNes_0.001WD_FalseTIED_1REP.csv',
    'training_metrics_PredNetBpD_9CLS_FalseNes_0.001WD_FalseTIED_2REP.csv',
    'training_metrics_PredNetBpD_10CLS_FalseNes_0.001WD_FalseTIED_2REP.csv'
]

# Colors for each model
colors = ['blue', 'green', 'red', 'purple', 'orange', 'brown', 'black', 'yellow', 'cyan', 'pink']

# Create a figure with four subplots
plt.figure(figsize=(20, 10))

# Train Loss subplot
plt.subplot(2, 2, 1)
for file, color in zip(model_files, colors):
    # Read the CSV file
    df = pd.read_csv(file, converters={
        'Train Accuracy': extract_numeric_value, 
        'Test Accuracy': extract_numeric_value
    })
    
    # Extract the number of circles from the filename
    num_cls = file.split('_')[3]
    
    # Plot train loss
    plt.plot(df['Epoch'], df['Train Loss'], label=f'Train Loss ({num_cls})', color=color)

plt.title('Train Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Test Loss subplot
plt.subplot(2, 2, 2)
for file, color in zip(model_files, colors):
    # Read the CSV file
    df = pd.read_csv(file, converters={
        'Train Accuracy': extract_numeric_value, 
        'Test Accuracy': extract_numeric_value
    })
    
    # Extract the number of circles from the filename
    num_cls = file.split('_')[3]
    
    # Plot test loss
    plt.plot(df['Epoch'], df['Test Loss'], label=f'Test Loss ({num_cls})', color=color)

plt.title('Test Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

# Train Accuracy subplot
plt.subplot(2, 2, 3)
for file, color in zip(model_files, colors):
    # Read the CSV file
    df = pd.read_csv(file, converters={
        'Train Accuracy': extract_numeric_value, 
        'Test Accuracy': extract_numeric_value
    })
    
    # Extract the number of circles from the filename
    num_cls = file.split('_')[3]
    
    # Plot train accuracy
    plt.plot(df['Epoch'], df['Train Accuracy'], label=f'Train Accuracy ({num_cls})', color=color)

plt.title('Train Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Test Accuracy subplot
plt.subplot(2, 2, 4)
for file, color in zip(model_files, colors):
    # Read the CSV file
    df = pd.read_csv(file, converters={
        'Train Accuracy': extract_numeric_value, 
        'Test Accuracy': extract_numeric_value
    })
    
    # Extract the number of circles from the filename
    num_cls = file.split('_')[3]
    
    # Plot test accuracy
    plt.plot(df['Epoch'], df['Test Accuracy'], label=f'Test Accuracy ({num_cls})', color=color)

plt.title('Test Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Adjust layout and save the plot
plt.tight_layout()
plt.savefig('circles_experiment.png', dpi=300)
plt.show()
```

```python
# Accuracy and Loss graphs around the learning rate changes for higher viewing resolution
lr_changes = [150, 225, 262]
    
for marked_epoch in lr_changes:
    print("Loss/Accuracy graph around learning rate change at epoch=" + str(marked_epoch))
    # Create a figure with four subplots
    plt.figure(figsize=(20, 10))
    
    # Train Loss subplot (epochs with learning rate changes)
    plt.subplot(2, 2, 1)
    for file, color in zip(model_files, colors):
        # Read the CSV file
        df = pd.read_csv(file, converters={
            'Train Accuracy': extract_numeric_value, 
            'Test Accuracy': extract_numeric_value
        })
        
        # Extract the number of circles from the filename
        num_cls = file.split('_')[3]
        
        # Select last 30 epochs
        subset = df.iloc[marked_epoch-15:marked_epoch+15]
        
        # Plot train loss
        plt.plot(subset['Epoch'], subset['Train Loss'], 
                 label=f'Train Loss ({num_cls})', color=color)
    
    plt.title('Train Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    # Test Loss subplot (last 30 epochs)
    plt.subplot(2, 2, 2)
    for file, color in zip(model_files, colors):
        # Read the CSV file
        df = pd.read_csv(file, converters={
            'Train Accuracy': extract_numeric_value, 
            'Test Accuracy': extract_numeric_value
        })
        
        # Extract the number of circles from the filename
        num_cls = file.split('_')[3]
        
        # Select last 30 epochs
        subset = df.iloc[marked_epoch-15:marked_epoch+15]
        
        # Plot test loss
        plt.plot(subset['Epoch'], subset['Test Loss'], 
                 label=f'Test Loss ({num_cls})', color=color)
    
    plt.title('Test Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    # Train Accuracy subplot (last 30 epochs)
    plt.subplot(2, 2, 3)
    for file, color in zip(model_files, colors):
        # Read the CSV file
        df = pd.read_csv(file, converters={
            'Train Accuracy': extract_numeric_value, 
            'Test Accuracy': extract_numeric_value
        })
        
        # Extract the number of circles from the filename
        num_cls = file.split('_')[3]
        
        # Select last 30 epochs
        subset = df.iloc[marked_epoch-15:marked_epoch+15]
        
        # Plot train accuracy
        plt.plot(subset['Epoch'], subset['Train Accuracy'], 
                 label=f'Train Accuracy ({num_cls})', color=color)
    
    plt.title('Train Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Test Accuracy subplot (last 30 epochs)
    plt.subplot(2, 2, 4)
    for file, color in zip(model_files, colors):
        # Read the CSV file
        df = pd.read_csv(file, converters={
            'Train Accuracy': extract_numeric_value, 
            'Test Accuracy': extract_numeric_value
        })
        
        # Extract the number of circles from the filename
        num_cls = file.split('_')[3]
        
        # Select last 30 epochs
        subset = df.iloc[marked_epoch-15:marked_epoch+15]
        
        # Plot test accuracy
        plt.plot(subset['Epoch'], subset['Test Accuracy'], 
                 label=f'Test Accuracy ({num_cls})', color=color)
    
    plt.title('Test Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Adjust layout and save the plot
    plt.tight_layout()
    plt.savefig('lr_change_'+ str(marked_epoch) +'_epochs_circles_experiment.png', dpi=300)
    plt.show()
```

```python
# Print header
print("\n==== Model Final Accuracies ====")
print("-" * 50)

# Store results in a list for potential further processing
results = []

# Process each model file
for file in model_files:
    # Read the CSV file
    df = pd.read_csv(file, converters={
        'Train Accuracy': extract_numeric_value, 
        'Test Accuracy': extract_numeric_value
    })
    
    # Extract the number of circles from the filename
    num_cls = file.split('_')[3]
    
    # Get final accuracies
    final_train_accuracy = df['Train Accuracy'].iloc[-1]
    final_test_accuracy = df['Test Accuracy'].iloc[-1]
    
    # Print results
    print(f"{num_cls} Classes:")
    print(f"  Final Train Accuracy: {final_train_accuracy:.2f}")
    print(f"  Final Test Accuracy:  {final_test_accuracy:.2f}")
    print()
    
    # Store results
    results.append({
        'Classes': num_cls,
        'Final Train Accuracy': final_train_accuracy,
        'Final Test Accuracy': final_test_accuracy
    })

# Optional: Create a pandas DataFrame for easy viewing
results_df = pd.DataFrame(results)
print("\nResults DataFrame:")
print(results_df)
```
